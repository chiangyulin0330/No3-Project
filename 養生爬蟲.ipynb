```python
#啟富達總攬爬爬爬
#因為是養生爬蟲 所以目前先爬2頁  ヽ（・∀・）ノ

from bs4 import BeautifulSoup
import requests
import re

page_url = 'http://www.cfd.tw/mod/news/index.php?y&pn={0}' #因要爬好幾頁，所以網址列不能寫死 下面要格式化
#print res.encoding


#因要爬好幾頁，SO 以下的設定要丟進迴圈 ，讓他每頁都執行一次
for page in range(0,2):
    res = requests.get(page_url.format(page)) 
    res.encoding = 'utf-8'
    soup = BeautifulSoup(res.text)
    entries  = soup.select(".listTable tr")
#因要爬好幾頁，SO 以上的設定要丟進迴圈 ，讓他每頁都執行一次
    
    #以下為抓標題和內容網址，可能是我抓的範圍有問題 所以有用正規化處理一下。
    for item in entries:
        td = item.find('td')
        a = item.find('a')
        if td is not None: 
            text = item.select('td')[1].text + item.select('td')[2].text
            m = re.search('.+',text)
            if m:
                print(text)
                print(a['href'])
                
#以上程式碼如有問題請找 建復組長(0934347071)  ｂ（￣▽￣）ｄ
```
